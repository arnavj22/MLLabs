```python
iris_path = "/content/drive/MyDrive/ML/Lab 7/iris.csv"
df = pd.read_csv(iris_path)
```

```python
def discretize(df):
  for i in list(df):
    if i != "variety":
      col = df[i]
      print(i)
      bins = pd.cut(col, 3, labels = ["1", "2", "3"]) # Discretizes to bins
      df[i] = bins
```

```python
discretize(df)
```

```python
y = df['variety']
X = df.drop(['variety'], axis = 1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train = np.array(X_train.to_numpy(),dtype='float64')
y_train = y_train.to_numpy()
X_test = np.array(X_test.to_numpy(),dtype='float64')
y_test = y_test.to_numpy()
_, y_train = np.unique(y_train, return_inverse=True)
_, y_test = np.unique(y_test, return_inverse=True)
```

```python
from mlxtend.classifier import OneRClassifier
oner = OneRClassifier()
oner.fit(X_train, y_train)
y_pred = oner.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Attribute:", list(X)[oner.feature_idx_])
print("Rules:", oner.prediction_dict_)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
