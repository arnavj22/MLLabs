```python
iris_path = "/content/drive/MyDrive/ML/Lab 7/iris.csv"
df = pd.read_csv(iris_path)
```

```python
def get_bounds(df):
  b = []
  for i in list(df):
    if i != "variety":
      col = df[i]
      bins = pd.cut(col, 3) # Discretizes to bins
      b.append(bins.unique())
  return b
def discretize(df):
  for i in list(df):
    if i != "variety":
      col = df[i]
      print(i)
      bins = pd.cut(col, 3, labels = ["1", "2", "3"]) # Discretizes to bins
      df[i] = bins
```

```python
b = get_bounds(df)
```

```python
discretize(df)
```

```python
y = df['variety']
X = df.drop(['variety'], axis = 1)
```

```python
print(y.unique())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
labels = []
max_acc = 0
best_rules = []
best_attribute = ""
for i in list(X_train):
  col = X_train[i]
  rules = []
  acc = 0
  for j in ["1", "2", "3"]:
    rows = y_train.loc[col == j]
    acc += rows.value_counts().max()
    rules.append(rows.value_counts()[rows.value_counts() == rows.value_counts().max()].index.values[0])
  if(acc > max_acc):
    best_rules = rules
    max_acc = acc
    best_attribute = i
```

```python
print(best_rules, max_acc / len(X_train), best_attribute)
```

```python
col = X_test[best_attribute]
test_acc = 0
preds = []
gts = []
for i,j in enumerate(list(col)):
  pred = best_rules[int(j) - 1]
  gt = y_test.iloc[i]
  preds.append(pred)
  gts.append(gt)
  if(pred == gt):
    test_acc += 1
ind = list(X_train).index(best_attribute)
ranges = b[ind]
print("Test Accuracy:", test_acc / len(X_test))
print("Attribute:", best_attribute, "\nRules:")
for i, j in zip(ranges, best_rules):
  print(i, "->", j)
print("\nConfusion Matrix:\n", pd.crosstab(pd.Series(gts, name = "Actual"), pd.Series(preds, name = "Predicted")))
```
