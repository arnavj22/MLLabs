```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets, neighbors
from mlxtend.plotting import plot_decision_regions
```

```python
x = []
y = []
for i in range(-10, 10, 1):
  x.append([i, i**2 - 50])
  y.append(1)
for i in range(-10, 10, 1):
  x.append([i, -1 * i**2 + 50])
  y.append(2)
```

```python
x = np.array(x)
y = np.array(y)
```

```python
k = 2
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters= k)
label = kmeans.fit_predict(x)
l0 = x[label == 0]
l1 = x[label == 1]
plt.scatter(l0[:,0] , l0[:,1] , color = 'red')
plt.scatter(l1[:,0] , l1[:,1] , color = 'black')
plt.title('Predicted Clustering')
plt.show()
l0 = x[y == 1]
l1 = x[y == 2]
plt.scatter(l0[:,0] , l0[:,1] , color = 'red')
plt.scatter(l1[:,0] , l1[:,1] , color = 'black')
plt.title('Ground Truth')
plt.show()
```

```python
from sklearn.datasets import make_blobs
x, y = make_blobs(n_samples=[50, 500, 50], n_features=2,centers= [(-5,2), (0,2), (5, 2)], random_state=0)
```

```python
k = 3

kmeans = KMeans(n_clusters= k)
label = kmeans.fit_predict(x)
l0 = x[label == 0]
l1 = x[label == 1]
l2 = x[label == 2]
plt.scatter(l0[:,0] , l0[:,1] , color = 'red')
plt.scatter(l1[:,0] , l1[:,1] , color = 'black')
plt.scatter(l2[:,0] , l2[:,1] , color = 'blue')
plt.title('Predicted Clustering')
plt.show()
l0 = x[y == 0]
l1 = x[y == 1]
l2 = x[y == 2]
plt.scatter(l0[:,0] , l0[:,1] , color = 'black')
plt.scatter(l1[:,0] , l1[:,1] , color = 'red')
plt.scatter(l2[:,0] , l2[:,1] , color = 'blue')
plt.title('Ground Truth')
plt.show()
```
